```python
# 2. Convert PDFs → page images
```

```python
from pdf2image import convert_from_path
from PIL import Image

Image.MAX_IMAGE_PIXELS = None  # disable decompression bomb limit

# Set this to your Poppler bin directory if needed, otherwise keep as None
POPPLER_PATH = None  # e.g. r"C:\tools\poppler-24.02.0\Library\bin"

pdf_dir = DATA_DIR / "data/printed_sources"
out_root = DATA_DIR / "data/printed_pages"
out_root.mkdir(parents=True, exist_ok=True)

print("PDF dir:", pdf_dir)
print("Output pages dir:", out_root)

for pdf_path in pdf_dir.glob("*.pdf"):
    print(f"\nProcessing PDF: {pdf_path.name}")
    source_name = pdf_path.stem
    out_dir = out_root / source_name
    out_dir.mkdir(parents=True, exist_ok=True)

    page_num = 1
    while True:
        try:
            pages = convert_from_path(
                str(pdf_path),
                dpi=200,                     # lower than 300 to save memory
                first_page=page_num,
                last_page=page_num,
                poppler_path=POPPLER_PATH,
            )
        except Exception as e:
            print(f"  Stopping at page {page_num}: {e}")
            break

        if not pages:
            break

        page = pages[0]
        out_path = out_dir / f"{source_name}_page_{page_num-1:03d}.png"
        page.save(out_path)
        print(f"  Saved {out_path.name}")
        page_num += 1

print("\nDone converting PDFs to page images.")
```

```python
# 4. Auto page→line dataset using pages_gt_auto.csv
```

```python
from tqdm import tqdm
import cv2

pages_root = DATA_DIR / "data/printed_pages"
lines_root = DATA_DIR / "data/lines"
lines_root.mkdir(parents=True, exist_ok=True)

auto_gt_path = DATA_DIR / "data/pages_gt_auto.csv"
print("Using auto ground truth CSV:", auto_gt_path)

if not auto_gt_path.exists():
    raise FileNotFoundError(f"Missing {auto_gt_path}. Please create pages_gt_auto.csv first.")

pages_gt = pd.read_csv(auto_gt_path)

print("Rows in pages_gt_auto:", len(pages_gt))

records = []  # will become lines_auto.csv: image_path,text

for _, row in tqdm(pages_gt.iterrows(), total=len(pages_gt)):
    page_rel = row["page_image_path"]
    text = str(row["text"])

    page_path = DATA_DIR / page_rel
    if not page_path.exists():
        print(f"WARNING: page image not found, skipping: {page_path}")
        continue

    try:
        bin_img = preprocess_page(page_path)
        main_img = crop_main_text(bin_img)
        line_imgs = segment_lines(main_img)
    except Exception as e:
        print(f"ERROR processing {page_path}: {e}")
        continue

    if len(line_imgs) == 0:
        continue

    for i, line_img in enumerate(line_imgs):
        line_path = lines_root / f"{page_path.stem}_line_{i:03d}.png"
        cv2.imwrite(str(line_path), line_img)
        # For now we attach the full page text (same as in your original notebook).
        records.append({"image_path": str(line_path), "text": text})

df_lines = pd.DataFrame(records)
lines_auto_csv = DATA_DIR / "data/lines_auto.csv"
df_lines.to_csv(lines_auto_csv, index=False)
print("Saved auto line dataset to:", lines_auto_csv)
print("Number of line samples:", len(df_lines))
```

```python
## 5. Split into train / val / test
```

```python
from sklearn.model_selection import train_test_split

df = pd.DataFrame(records)
train_df, tmp = train_test_split(df, test_size=0.3, random_state=42)
val_df, test_df = train_test_split(tmp, test_size=0.5, random_state=42)

out_data_dir = DATA_DIR / "data"
train_df.to_csv(out_data_dir / "lines_train.csv", index=False)
val_df.to_csv(out_data_dir / "lines_val.csv", index=False)
test_df.to_csv(out_data_dir / "lines_test.csv", index=False)

train_df.head()
```
